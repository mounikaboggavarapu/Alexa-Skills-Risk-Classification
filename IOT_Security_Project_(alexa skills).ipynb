```python
# import needed libraries
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
nltk.download("punkt")  # for sentence splitting
import nltk
nltk.download('punkt_tab')


from sentence_transformers import SentenceTransformer, util

```

```python
# recognizing categories
data_categories = {
    "name": [
        "name",
        "full name",
        "nickname",
        "username",
        "first name",
        "last name"
    ],
    "email": [
        "email",
        "email address"
    ],
    "phone_number": [
        "phone number",
        "telephone number",
        "mobile number",
        "cell number"
    ],
    "location": [
        "location",
        "gps data",
        "address",
        "geolocation",
        "coordinates"
    ],
    "payment_info": [
        "credit card",
        "payment information",
        "billing address",
        "bank account",
        "financial information"
    ],
    "device_usage": [
        "device usage",
        "usage data",
        "app usage",
        "interaction data"
    ],
    "health_data": [
        "health data",
        "health information",
        "medical data",
        "fitness data",
        "heart rate"
    ],
    "personal_identifiers": [
        "personal identifier",
        "unique id",
        "account details",
        "account information",
        "user id"
    ]
}
```

```python
# get embeddings for all the category phrases using the model
model = SentenceTransformer("all-MiniLM-L6-v2")

category_phrases = []
for category, synonyms in data_categories.items():
    for phrase in synonyms:
        category_phrases.append((category, phrase))

# Just the phrase text for embedding
category_texts = [pair[1] for pair in category_phrases]
category_embeddings = model.encode(category_texts, convert_to_numpy=True)
```

```python
# splits the input text into sentences using nltk
def split_into_sentences(text):
    """
    Splits text into sentences using nltk's sentence tokenizer.
    """
    if not text or not isinstance(text, str):
        return []
    return sent_tokenize(text)

# checks which categories show up in the text based on sentence similarity
def detect_data_categories(text, threshold=0.4):
    """
    Returns a set of category names that appear in the text
    based on BERT embeddings and a similarity threshold.
    """
    # Break text into sentences (chunks)
    chunks = split_into_sentences(text)
    if not chunks:
        return set()

    # Embed each sentence
    chunk_embeddings = model.encode(chunks, convert_to_numpy=True)

    matched_categories = set()

    # For each chunk, find the phrase with highest similarity
    for chunk_emb in chunk_embeddings:
        sims = util.cos_sim(chunk_emb, category_embeddings)[0].tolist()  # similarity to each category phrase
        max_score = max(sims)
        max_index = sims.index(max_score)

        if max_score >= threshold:
            cat = category_phrases[max_index][0]  # the category name
            matched_categories.add(cat)

    return matched_categories
```

```python
# run the category detection on both skill descriptions and privacy policies with different thresholds
df["Categories_in_Description"] = df["Skill_Description"].apply(
    lambda x: detect_data_categories(x, threshold=0.2)
)
df["Categories_in_Policy"] = df["Privacy_Policy_Combined"].apply(
    lambda x: detect_data_categories(x, threshold=0.3)
)
```

```python
# set up functionality categories with synonyms and needed data, then embed all the synonym phrases
functionality_dict = {
    # Cameras & doorbells
    "home_security": {
        "synonyms": [
            "camera", "doorbell", "security camera",
            "surveillance", "live stream", "stream", "monitor"
        ],
        "needed_data": {"video_data", "device_usage", "location"}
    },
    # Thermostats & climate
    "thermostat_control": {
        "synonyms": [
            "thermostat", "temperature", "climate control",
            "heating", "cooling"
        ],
        "needed_data": {"temperature_data", "device_usage", "location"}
    },
    # Smart lighting
    "lighting_control": {
        "synonyms": [
            "light", "lamp", "lighting", "bulb"
        ],
        "needed_data": {"device_usage"}
    },
    # Smart plugs and outlets
    "plug_control": {
        "synonyms": [
            "plug", "outlet", "socket"
        ],
        "needed_data": {"device_usage"}
    },
    # Everything else that’s “smart home” control
    "generic_home_automation": {
        "synonyms": [
            "automate", "control", "home automation", "smart home"
        ],
        "needed_data": {"device_usage", "location"}
    }
}



model = SentenceTransformer("all-MiniLM-L6-v2")

functionality_phrases = []
for func, details in functionality_dict.items():
    for syn in details["synonyms"]:
        functionality_phrases.append((func, syn))

functionality_texts = [fp[1] for fp in functionality_phrases]
functionality_embs = model.encode(functionality_texts, convert_to_numpy=True)

```

```python
# figure out the most likely functionality from the description using similarity with known phrases
def detect_functionality(description, threshold=0.3):
    sents = sent_tokenize(description)
    if not sents:
        return "unknown"

    desc_embs = model.encode(sents, convert_to_numpy=True)

    best_func = "unknown"
    best_score = 0.0
    for i, phrase_text in enumerate(functionality_texts):
        max_sim = 0.0
        for emb in desc_embs:
            sim = util.cos_sim(emb, functionality_embs[i])[0][0].item()
            if sim > max_sim:
                max_sim = sim

        if max_sim > best_score:
            best_score = max_sim
            best_func = functionality_phrases[i][0]

    return best_func if best_score >= threshold else "unknown"
```

```python
# detect functionality from each skill description and show a few results
df["Detected_Functionality"] = df["Skill_Description"].apply(detect_functionality)
df[["Skill_Name", "Skill_Description", "Detected_Functionality"]].head()
```

```python
def evaluate_data_need(row):
    func = row["Detected_Functionality"]
    desc_cats = row["Categories_in_Description"]
    policy_cats = row["Categories_in_Policy"]

    # The categories the skill is actually collecting (union of desc & policy)
    all_collected = desc_cats.union(policy_cats)

    # If functionality is unknown, we can’t do a definitive check
    if func == "unknown":
        return {
            "needed_for_func": set(),
            "unnecessary": all_collected  # we can't confirm any are needed
        }

    # The categories typical for that functionality
    needed_for_func = functionality_dict[func]["needed_data"]

    # “Unnecessary” data are those in all_collected but not in needed_for_func
    unnecessary = all_collected - needed_for_func

    return {
        "needed_for_func": needed_for_func,
        "unnecessary": unnecessary
    }

df["Need_Evaluation"] = df.apply(evaluate_data_need, axis=1)

# You can split the result into separate columns if you prefer
df["Needed_Data"] = df["Need_Evaluation"].apply(lambda x: x["needed_for_func"])
df["Unnecessary_Data"] = df["Need_Evaluation"].apply(lambda x: x["unnecessary"])

df[["Skill_Name", "Needed_Data", "Unnecessary_Data"]]
```

```python
import pandas as pd
import itertools
import numpy as np
from scipy.spatial.distance import jensenshannon


sensitive_set = {"health_data", "payment_info", "phone_number", "location", "video_data"}

def risk_from_weights(row, w):
    s = w["sens"] * sum(cat in sensitive_set for cat in row["Categories_in_Policy"])
    s += w["unnec"] * len(row["Unnecessary_Data"])
    s += w["disc"]  * len(row["Missing_in_Description"])
    s += w["rev"]   * (row["Number_of_review"] > 5000)
    s += w["rate"]  * (row["ratings"] > 3.5)
    return s

def compute_distribution(weights):
    # apply scoring + bucket into low/medium/high
    scores = df.apply(lambda r: risk_from_weights(r, weights), axis=1)
    labels = scores.apply(lambda s: "low" if s <= 3 else ("medium" if s <= 5 else "high"))
    dist = labels.value_counts(normalize=True).reindex(["low","medium","high"], fill_value=0)
    return dist.values  # array([p_low, p_med, p_high])

# define baseline weights and distribution
baseline = {"sens":2, "unnec":0.5, "disc":0.5, "rev":-1, "rate":-2}
baseline_dist = compute_distribution(baseline)

# set up grid of candidate weight tuples
param_grid = {
    "sens": [1, 2, 3],
    "unnec":[0.2, 0.5, 1.0],
    "disc": [0.2, 0.5, 1.0],
    "rev":  [0, -1],
    "rate": [0, -1, -2]
}

# exhaustively evaluate each combination
best = None
best_score = float("inf")

for comb in itertools.product(*(param_grid[k] for k in param_grid)):
    w = dict(zip(param_grid.keys(), comb))
    dist = compute_distribution(w)
    # Jensen–Shannon divergence is symmetric; square it to get a true metric
    js = jensenshannon(baseline_dist, dist)**2
    if js < best_score:
        best_score = js
        best = w

print("Baseline distribution:", dict(zip(["low","medium","high"], baseline_dist)))
print("Best weights:", best)
print("Minimal JS divergence:", best_score)
print("Resulting distribution:", dict(zip(["low","medium","high"], compute_distribution(best))))
```

```python

best_weights = {
    "sens": 2,
    "unnec": 0.5,
    "disc": 0.5,
    "rev": 0,
    "rate": -2
}

# scoring function using those weights
def compute_score(row):
    s = best_weights["sens"] * sum(cat in sensitive_set for cat in row["Categories_in_Policy"])
    s += best_weights["unnec"] * len(row["Unnecessary_Data"])
    s += best_weights["disc"]  * len(row["Missing_in_Description"])
    s += best_weights["rev"]   * (row["Number_of_review"] > 5000)
    s += best_weights["rate"]  * (row["ratings"] > 3.5)
    return s

# apply to get a numeric Risk_Score
df["Risk_Score"] = df.apply(compute_score, axis=1)

# map numeric scores to your three risk buckets
def map_level(score):
    if score <= 4:
        return "low"
    elif score <= 8:
        return "medium"
    else:
        return "high"

df["Risk_Level"] = df["Risk_Score"].apply(map_level)

df.to_csv("risk_scored_skills.csv", index=False)

# quick sanity check
print(df[["Skill_Name", "Risk_Score", "Risk_Level"]].head())
```

```python
import pandas as pd
import re


# define a mapping from canonical cert name → list of synonyms
cert_map = {
    "CCPA": [
        "CCPA",
        "California Consumer Privacy Act"
    ],
    "GDPR": [
        "GDPR",
        "General Data Protection Regulation"
    ],
    "HIPAA": [
        "HIPAA",
        "Health Insurance Portability and Accountability Act"
    ],
    "SOC 2": [
        "SOC 2",
        "Service Organization Control 2"
    ],
    "ISO 27001": [
        "ISO[- ]?27001",
        "International Organization for Standardization 27001"
    ],
    "ISO 27018": [
        "ISO[- ]?27018",
        "International Organization for Standardization 27018"
    ],
    "FCC": [
        "FCC",
        "Federal Communications Commission"
    ],
    "UL": [
        "UL",
        "Underwriters Laboratories"
    ],
    "CE": [
        "CE",
        "Conformité Européenne"
    ],
    "CSA": [
        "CSA",
        "Canadian Standards Association"
    ],
    "Energy Star": [
        "Energy Star"
    ],
    "Wi-Fi CERTIFIED": [
        "Wi[- ]?Fi[- ]?CERTIFIED",
        "Wi[- ]?Fi Alliance"
    ],
    "Zigbee Certified": [
        "Zigbee[- ]?Certified",
        "Connectivity Standards Alliance"
    ],
    "Thread Certified": [
        "Thread[- ]?Certified"
    ],
    "Matter Certified": [
        "Matter[- ]?Certified"
    ],
    "Bluetooth Certified": [
        "Bluetooth[- ]?Certified",
        "Bluetooth Special Interest Group"
    ]
}

# pre-compile one regex per canonical certification
cert_patterns = {
    canon: re.compile(r"\b(?:" + "|".join(syns) + r")\b", re.IGNORECASE)
    for canon, syns in cert_map.items()
}

# extraction function: return only canonical names
def extract_certifications(text):
    if not isinstance(text, str) or not text:
        return []
    found = []
    for canon, pat in cert_patterns.items():
        if pat.search(text):
            found.append(canon)
    return found

# apply to your combined privacy-policy column
df["Certifications"] = df["Privacy_Policy_Combined"].apply(extract_certifications)
from IPython.display import display
display(df[["Skill_Name", "Certifications"]])
```
